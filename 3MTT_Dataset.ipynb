```python
#Import Libraries and Modules needed
import requests  # Library for making HTTP requests
import pandas as pd  # Library for data manipulation and analysis
import numpy as np  # Library for numerical computing
import ast  # Module for literal string evaluation
import geopandas as gpd  # Library for working with geospatial data
import matplotlib.pyplot as plt  # Library for creating visualizations
import seaborn as sns  # Library for statistical data visualization
```

```python
# Initialize an empty list to store data from each page
all_results = []

# Loop through pages from 1 to 739
for page in range(1, 740):
    #  URL for each page
    url = f"https://app.3mtt.training/selected/fellows?page={page}"

    # Send GET URL endpoint
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Extract JSON data from the response
        data = response.json()

        # Extract the 'results' data from the 'data' column
        results_data = data['data']['results']

        # Append results_data to all_results list
        all_results.extend(results_data)

        # Print progress
        print(f"Page {page} processed. Total records: {len(all_results)}")

    else:
        print(f"Failed to fetch data from page {page}. Status code: {response.status_code}")

# Convert all_results to a DataFrame
df = pd.DataFrame(all_results)

# Save the DataFrame as a CSV file
df.to_csv('3MTT_Dataset.csv', index=False)

print("Scraping and saving completed.")
```

```python
# Load the CSV file
df = pd.read_csv('3MTT_Dataset.csv')

# Convert string representations of dictionaries back to dictionaries
df['cohort'] = df['cohort'].apply(ast.literal_eval)

# Convert the nested "cohort" column into separate columns
df['cohort_id'] = df['cohort'].apply(lambda x: x['id'])
#df['cohort_name'] = df['cohort'].apply(lambda x: x['name'])
df['cohort_name'] = df['cohort'].apply(lambda x: x['rank'])
df['start_date'] = df['cohort'].apply(lambda x: x['start_date'])
df['end_date'] = df['cohort'].apply(lambda x: x['end_date'])
df['status'] = df['cohort'].apply(lambda x: x['status'])

# Drop the original 'cohort' column
df.drop('cohort', axis=1, inplace=True)

# Save the modified DataFrame as a new CSV file
df.to_csv('3MTT_Dataset_Modified.csv', index=False)

print("Modified CSV file saved as '3MTT_Dataset_Modified.csv'")
```

```python
# Get the total count of residences by state
residence_counts = df['residence'].value_counts()

# Load the shapefile containing the boundaries of Nigerian states
nigeria_map = gpd.read_file('STATES/nigeria_shapefile.shp')

# Merge the shapefile with the total count of residences by state
nigeria_map = nigeria_map.merge(residence_counts, left_on='NAME_1', right_index=True)

# Plotting
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
nigeria_map.plot(column='residence', cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# Annotate each state with its count
for idx, row in nigeria_map.iterrows():
    ax.annotate(text=str(row['residence']), xy=row['geometry'].centroid.coords[0], color='black', fontsize=8, ha='center')

ax.set_title('Total Count of Residences by State on Nigerian Map')
plt.show()
```

```python
#Heat Map
# Group by 'residence' and 'current_learning_course', then count the occurrences
course_counts_by_state = df.groupby(['residence', 'current_learning_course']).size().reset_index(name='count')

# Pivot the DataFrame for easier plotting
pivot_df = course_counts_by_state.pivot(index='residence', columns='current_learning_course', values='count')

# Rearrange the columns within each row from highest to lowest count
pivot_df = pivot_df.apply(lambda x: x.sort_values(ascending=False), axis=1)

# Plot using Seaborn
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_df, cmap='coolwarm', annot=True, fmt="g", linewidths=.5)
plt.title('Total Count of Each Current Learning Course by Each State (Highest to Lowest)')
plt.xlabel('Current Learning Course', horizontalalignment='right')
plt.ylabel('State')
plt.xticks(rotation=45, horizontalalignment='right')
plt.tight_layout()
plt.show()
```
